```python
import cv2
from google.colab.patches import cv2_imshow # Import the Colab patch for imshow

# Load an image from file, verify the path is correct
img_path = '/content/sample_data/husky-gallery-outdoors-1-min.jpg'
imgrgb = cv2.imread(img_path)

# Check if the image was loaded successfully
if imgrgb is None:
  print(f"Error: Could not load image from {img_path}. Check if the file exists and the path is correct.")
else:
  cv2_imshow(imgrgb) # Display the image using the Colab patch
```

```python
from matplotlib import pyplot as plt #plot graphs
import numpy as np #arrays and matrics
```

```python
imgrgb.shape
```

```python
imggray = cv2.cvtColor(imgrgb, cv2.COLOR_BGR2GRAY) #convert to gray
```

```python
imggray.shape
```

```python
cv2_imshow(imggray)
```

```python
#Define verticle kernal
v_kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
#Define Horizontal kernal
h_kernel = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])
```

```python
v_image = cv2.filter2D(imggray, -1, v_kernel) #filtering
h_image = cv2.filter2D(imggray, -1, h_kernel)
```

```python
cv2_imshow(v_image)
```

```python
cv2_imshow(h_image)
```

```python
plt.imshow(imggray, cmap='gray')
plt.title('Image')

plt.figure()
plt.imshow(v_image, cmap='gray')
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image, cmap='gray')
plt.title('Horizontal Image')
```
